```python
import numpy as np
import pandas as pd
```

```python
# Thông tin về dữ liệu
info = pd.read_csv("../ml-100k/u.info", header=None, sep="\t")
info.head()
```

```python
# Các loại occupation
occupation = pd.read_csv("../ml-100k/u.occupation", header=None, sep="\t")
print("Số lượng ngành nghề:")
print(len(occupation))
print("Các loại ngành nghề:")
print(occupation[0].tolist())
```

```python
# Các thể loại phim
genre = pd.read_csv("../ml-100k/u.genre", header=None, sep="\t")
print("Số lượng thể loại phim:")
print(len(genre))
print("Các thể loại phim:")
print(genre[0].tolist())
```

```python
# Dữ liệu về người dùng
users = pd.read_csv("../ml-100k/u.user", header=None, sep="|")
users.columns = ['user_id', 'age', 'gender', 'occupation', 'zip_code']
users.head()
```

```python
# Dữ liệu về phim
movies = pd.read_csv("../ml-100k/u.item", header=None, sep="|", encoding='latin-1')
movies.columns = ['movie_id', 'title', 'release_date', 'video_release_date', 'IMDb_URL'] + genre[0].tolist()
movies.head()
```

```python
# Dữ liệu về Ratings
ratings = pd.read_csv("../ml-100k/u.data", header=None, sep="\t")
ratings.columns = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings.head()
```

```python
# Gộp dữ liệu người dùng và ratings
user_ratings = pd.merge(ratings, users, on='user_id')
user_ratings.head()
```

```python
# Top 10 phim được đánh giá cao nhất
top_movies = user_ratings.groupby('movie_id')['rating'].mean().sort_values(ascending=False).head(10)
top_movies = pd.merge(top_movies, movies[['movie_id', 'title']], on='movie_id')
print("Top 10 phim được đánh giá cao nhất:")
print(top_movies[['title', 'rating']])
```

```python
# Danh sách thể loại phim
genre_list = genre[0].tolist()

# Gộp dữ liệu ratings và movies
data_item = pd.merge(
    ratings[['user_id', 'movie_id', 'rating']],
    movies,
    on='movie_id'
 )

# Loại bỏ các cột không cần thiết
data_item.drop(
    columns=['user_id', 'movie_id', 'release_date', 'video_release_date', 'IMDb_URL'],
    inplace=True
 )

# Với mỗi thể loại, lấy top 3 phim theo rating trung bình
top_3_genre = pd.DataFrame()
for gen in genre_list:
    g_r = data_item[data_item[gen] == 1]
    new_gen = pd.DataFrame(
        g_r.groupby('title', as_index=False)['rating']
        .mean()
        .sort_values(['rating', 'title'], ascending=[False, True])
        .head(3)
    )
    new_gen.insert(0, 'genre', gen)
    top_3_genre = pd.concat([top_3_genre, new_gen], ignore_index=True)

top_3_genre
```

```python
# Gộp dữ liệu ratings và users
data_user_og = pd.merge(ratings, users, on='user_id')
data_user_og.drop(columns=['user_id'], inplace=True)

# Gộp dữ liệu ratings, users và movies
data_user_item = pd.merge(data_user_og, movies, on='movie_id')
data_user_item.drop(columns=['movie_id', 'release_date', 'video_release_date', 'IMDb_URL'], inplace=True)
top_3_genre = pd.DataFrame()
for occ in occupation[0].tolist():
    o_r = data_user_item[data_user_item['occupation'] == occ]
    new_occ = pd.DataFrame(
        o_r.groupby('title', as_index=False)['rating']
        .mean()
        .sort_values(['rating', 'title'], ascending=[False, True])
        .head(3)
    )
    new_occ.insert(0, 'occupation', occ)
    top_3_genre = pd.concat([top_3_genre, new_occ], ignore_index=True)

top_3_genre
```

```python
# Tạo cột nhóm tuổi
bins = [0, 6, 12, 18, 25, 35, 45, 50, 56, 100]
labels = ['0-5', '6-11', '12-17', '18-24', '25-34', '35-44', '45-49', '50-55', '56+']
users['age_group'] = pd.cut(users['age'], bins=bins, labels=labels, right=False)

# Gộp ratings với users để lấy thông tin nhóm tuổi
data_user_age = pd.merge(ratings, users, on='user_id')
data_user_age.drop(columns=['user_id'], inplace=True)

# Gộp tiếp với movies để lấy thông tin phim
data_user_item_age = pd.merge(data_user_age, movies, on='movie_id')
data_user_item_age.drop(columns=['movie_id', 'release_date', 'video_release_date', 'IMDb_URL'], inplace=True)

# Lấy top 3 phim theo rating trung bình cho từng nhóm tuổi
data_user_item_age_sorted = (
    data_user_item_age
    .groupby(['age_group', 'title'], as_index=False)['rating']
    .mean()
    .sort_values('rating', ascending=False)
 )

top_3_age = (
    data_user_item_age_sorted
    .groupby('age_group')
    .head(3)
    .sort_values(['age_group', 'title'], ascending=[True, True])
    .reset_index(drop=True)
 )

top_3_age
```
